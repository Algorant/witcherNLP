```python
import requests
from bs4 import BeautifulSoup

url = "https://awoiaf.westeros.org/index.php/List_of_characters"
request = requests.get(url)
soup = BeautifulSoup(request.text, 'html.parser')

urls = []
for link in soup.find_all('a'):
    print(link.get('href'))
```
